```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Load the dataset
df_climate = pd.read_csv('Climate Change.csv')
```

```python
df_climate.columns
```

```python
#Info of the data
df_climate.info()
```

```python
#Describe the dataset
df_climate.describe().T
```

```python
# Convert 'Year' and 'Month' to a datetime object
df_climate['date'] = pd.to_datetime(df_climate[['Year', 'Month']].assign(day=1))
```

```python
import matplotlib.pyplot as plt

# List of variables to plot
variables = ['CO2', 'MEI', 'CH4', 'N2O', 'CFC-11', 'CFC-12', 'TSI', 'Aerosols', 'Temp']

# Plotting multiple time series on the same plot
plt.figure(figsize=(14, 8))

# Plot each variable
for var in variables:
    plt.plot(df_climate['date'], df_climate[var], marker='o', linestyle='-', label=var)

# Adding titles and labels
plt.title('Trends of Multiple Climate Variables Over Time', fontsize=16)
plt.xlabel('Date', fontsize=14)
plt.ylabel('Values', fontsize=14)
plt.legend(loc='upper right')
plt.grid(True)

# Display the plot
plt.show()
```

```python
# Plotting the line chart
plt.figure(figsize=(10, 6))
plt.plot(df_climate['date'], df_climate['CO2'], marker='o', linestyle='-', color='b')

# Adding titles and labels
plt.title('CO2 Concentration Over Time', fontsize=16)
plt.xlabel('Date', fontsize=14)
plt.ylabel('CO2 Concentration', fontsize=14)
plt.grid(True)

# Display the plot
plt.show()
```
